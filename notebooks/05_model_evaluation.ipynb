```python
import sys, os
import numpy as np
repo_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(os.path.join(repo_root, "src"))

from evaluate_nn import evaluate_nn
```

```python
# Load embeddings
X_train_tfidf = np.load("../data/embedded/tfidf/X_train.npy", allow_pickle=True)
X_val_tfidf   = np.load("../data/embedded/tfidf/X_val.npy", allow_pickle=True)
X_test_tfidf  = np.load("../data/embedded/tfidf/X_test.npy", allow_pickle=True)

X_train_sbert = np.load("../data/embedded/sbert/X_train.npy", allow_pickle=True)
X_val_sbert   = np.load("../data/embedded/sbert/X_val.npy", allow_pickle=True)
X_test_sbert  = np.load("../data/embedded/sbert/X_test.npy", allow_pickle=True)

# Load labels
y_train = np.load("../data/processed/y_train.npy", allow_pickle=True)
y_val   = np.load("../data/processed/y_val.npy", allow_pickle=True)
y_test  = np.load("../data/processed/y_test.npy", allow_pickle=True)
```

```python
# ---------------- TF-IDF Evaluation ----------------
model_tfidf, hist_tfidf = evaluate_nn(
    X_train_tfidf, y_train, X_val_tfidf, y_val, X_test_tfidf, y_test,
    hidden_units=[128],
    dropout_rate=0.3,
    learning_rate=0.0001,
    batch_size=32,
    optimizer="adam",
    embedding_name="tfidf",
    epochs=30
)
```

```python
model_sbert, hist_sbert = evaluate_nn(
    X_train_sbert, y_train, X_val_sbert, y_val, X_test_sbert, y_test,
    hidden_units=[256,128],
    dropout_rate=0.1,
    learning_rate=0.0001,
    batch_size=32,
    optimizer="adam",
    embedding_name="sbert",
    epochs=30
)
```
